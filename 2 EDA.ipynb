```python
import pickle

with open('astroworld.pkl', 'rb') as f:
    corpus_df = pickle.load(f)
```

```python
corpus_df.lyrics
```

```python
corpus_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
cv = CountVectorizer()
```

```python
lyrics_cv = cv.fit_transform(corpus_df.lyrics)
```

```python
dtm = pd.DataFrame(lyrics_cv.toarray())
dtm.columns = cv.get_feature_names()
dtm.index = corpus_df.index
```

```python
dtm
```

```python
tdm = dtm.transpose()
```

```python
tdm
```

```python
for i in tdm:
    print(i, '\n---')
    print(tdm[i].sort_values(ascending=False).head(5), '\n\n')
```

```python
top_words_dict = {}

for col in tdm.columns:
    top_words = tdm[col].sort_values(ascending=False).head(5)
    #top_words_dict[col] = list(top_words.values)
    top_words_dict[col] = list(zip(top_words.index, top_words.values))
```

```python
top_words_dict
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc = WordCloud(
    background_color='white',
    max_font_size=150,
    colormap='Dark2',
    collocations=False
)
```

```python
plt.rcParams['figure.figsize'] = [16, 6]
```

```python
import os

def generate_wordcloud(song_name):
    idx = list(tdm.columns).index(song_name)
    
    wc.generate(corpus_df.lyrics[song_name])
    
    plt.subplot(1, 1, 1)
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(corpus_df.index[idx])
    
    try: os.mkdir('wordclouds')
    except FileExistsError: pass
    
    plt.savefig(f"wordclouds/{''.join(song_name.split('.'))}")
    
    return plt.show()
```

```python
for song_name in tdm:
    generate_wordcloud(song_name)
```

### Topic Modeling

Input to Topic modeling model is a document-term matrix.

Each topic consists of a set of words where order doesn't matter.

LDA, Latent Dirichlet Allocation is a popular topic modeling technique.

Topic modeling is a form of:
- dimensionality reduction
- unsupervised learning

```python
from gensim import models, matutils
import scipy.sparse
```

```python
tdm
```

```python
sparse_counts = scipy.sparse.csc_matrix(tdm)
corpus = matutils.Sparse2Corpus(sparse_counts)
```

```python
id2word = dict((v, k) for k, v in cv.vocabulary_.items())
```

```python
lda = models.LdaModel(
    corpus=corpus, 
    id2word=id2word, 
    num_topics=3, 
    passes=10
)
```

```python
lda.print_topics(num_words=6)
```
